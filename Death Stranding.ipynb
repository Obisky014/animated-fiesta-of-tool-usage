```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import random
```

```python
# creating a theoretical table, because it is really hard to get the actual data
# List of death stranding items (Non-Lethal ones at least)

Items = [
    "Ladder", "Bridges Boots", "MULE Boots", "Container Repair Spray", "Climbing Anchor", "Blood Bag", "PCC", "Smoke-Emitting Decoy Cargo", "Vog-Emitting Decoy Cargo", "Power Skeleton", "Thermal Pad", "Speed Skeleton", "All-Terrain Skeleton", "Reverse Trike", "Truck"]

# Generate synthetic data

Data = {
    "Item Name":[],
    "Usage Count":[],
    "Weight (kg)":[],
    "Durability(%)":[],
}

# Generate random weight and durability values
item_weights = {
    "Ladder": 5.0, "Climbing Anchor": 3.0, "Container Repair Spray": 1.0, "Blood Bag": 0.8,
    "PCC": 2.5, "Smoke-Emitting Decoy Cargo": 4.0,
    "Vog-Emitting Decoy Cargo": 4.2,
    "Power Skeleton": 8.0, "Speed Skeleton": 6.0, "All-Terrain Skeleton": 6.0,
    "Thermal Pad": 1.2,
    "Reverse Trike": 50.0, "Truck": 200.0,
    "Bridges Boots": 0.5, "MULE Boots": 0.5
}

for item in Items:
    for _ in range(random.randint(5, 15)):  # Each item appears multiple times with different values
        Data["Item Name"].append(item)
        Data["Usage Count"].append(random.randint(10, 500))  # Random usage count
        Data["Weight (kg)"].append(item_weights[item])  # Assign weight
        Data["Durability(%)"].append(random.randint(50, 100))  # Random durability percentage

# Create DataFrame
df = pd.DataFrame(Data)

# Save dataset
df.to_csv("death_stranding_items.csv", index=False)

# Display first few rows
df.head()

```

```python
# general info look
df.info()
```

```python
# visualizing with a scatterplot. Redundant, but i still wanted to do it anyways
plt.figure(figsize=(10,7))
plt.scatter(df["Item Name"], df["Usage Count"], color='cyan')
plt.xlabel("Item Name")
plt.ylabel("Usage Count")
plt.title("Item Usage")
plt.xticks(rotation=90)
plt.show()
```

```python
# bar chart visualization
plt.figure(figsize=(10,7))
plt.bar(df["Item Name"], df["Usage Count"], color='cyan')
plt.xlabel("Item Name")
plt.ylabel("Usage Count")
plt.title("Item Usage")
plt.xticks(rotation=90)
plt.show()
```

```python
# What's better than having a monochromatic set of colours describing usage? Multiple colours describing Usage. So we assign different colors
colors = plt.cm.viridis(np.linspace(0, 1, len(df)))  # Generates distinct colors

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(df["Item Name"], df["Usage Count"], color=colors)

# Rotate x labels for better readability
plt.xticks(rotation=45)
plt.xlabel("Item Name")
plt.ylabel("Usage Count")
plt.title("Death Stranding Item Usage but with Different Colors")

plt.show()
```
